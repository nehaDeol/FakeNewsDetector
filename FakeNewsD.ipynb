```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.feature_extraction.text import CountVectorizer 
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn import feature_extraction, linear_model, model_selection, preprocessing
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
```

```python
dataF= pd.read_csv("Fake.csv")
dataT=pd.read_csv("True.csv")
```

```python
dataF.shape
```

```python
dataT.shape
```

```python
# Add flag to track fake and real
dataF['target']= 'fake'
dataT['target']= 'true'
```

```python
# concatenate dataframes 
data = pd.concat([dataF, dataT]).reset_index(drop = True)
data.shape
```

```python
from sklearn.utils import shuffle
data= shuffle(data)
data= data.reset_index(drop = True)
```

```python
# check the data 
data.head()
```

```python
# removing column not needed aka date
data.drop(["date"], axis=1, inplace=True)
data.head()
```

```python
# removing title column
data.drop(["title"], axis=1, inplace=True)
data.head()
```

```python
# convert all to lowercase
data['text']= data['text'].apply(lambda x: x.lower())
data.head()
```

```python
import string 
def punctuation_removal(text):
    all_list = [char for char in text if char not in string.punctuation]
    clean_str= ''.join(all_list)
    return clean_str
data['text']= data['text'].apply(punctuation_removal)
data.head()
```

```python
# removing stopwords
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
stop = set(stopwords.words('english'))

# import nltk
# nltk.download('stopwords')

data['text']= data['text'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop)]))
```

```python
# most frequent word counter
from nltk import tokenize 
token_space = tokenize.WhitespaceTokenizer()

def counter(text, column_text, quantity):
    all_words = ' '.join([text for text in text[column_text]])
    token_phrase = token_space.tokenize(all_words)
    frequency = nltk.FreqDist(token_phrase)
    df_frequency = pd.DataFrame({"Word": list(frequency.keys()),
                                "Frequency": list(frequency.values())})
    df_frequency = df_frequency.nlargest(columns = "Frequency", n = quantity)
    plt.figure(figsize=(12,8))
    ax = sns.barplot(data = df_frequency, x = "Word", y = "Frequency", color = 'blue')
    ax.set(ylabel = "Count")
    plt.xticks(rotation='vertical')
    plt.show()
```

```python
# most frequent words in fake news
counter(data[data["target"] == "fake"], "text", 20)
```

```python
# most frequent words in true news
counter(data[data["target"] == "true"], "text", 20)
```

```python
# function to plot the confusion matrix
from sklearn import metrics
import itertools 

def plot_confusion_matrix(cm, classes,
                         normalize=False,
                         title='Confusion matrix',
                         cmap=plt.cm.Blues):
    
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)
    
    if normalize: 
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('Confusion matriz, without normalization')
        
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, cm[i, j],
                horizontalalignment="center",
                color="white" if cm[i, j] > thresh else "black")
        
    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    
```

```python
# split the data set
X_train, X_test, y_train, y_test = train_test_split(data['text'], data.target, test_size=0.2, random_state=42)
```

```python
X_train.head()
# y_test.tail()
```

```python
from sklearn.tree import DecisionTreeClassifier 

# Vectorizing and applying TF-IDF
pipe = Pipeline([('vect', CountVectorizer()),
                ('tfidf', TfidfTransformer()),
                ('model', DecisionTreeClassifier(criterion= 'entropy',
                                                max_depth = 20,
                                                splitter='best',
                                                random_state=42))])

# fitting the model 
model = pipe.fit(X_train, y_train)

# accuracy
prediction = model.predict(X_test)
print("accuracy: {}%".format(round(accuracy_score(y_test, prediction)*100, 2)))
```

```python
cm = metrics.confusion_matrix(y_test, prediction)
plot_confusion_matrix(cm, classes=['Fake', 'Real'])
```
